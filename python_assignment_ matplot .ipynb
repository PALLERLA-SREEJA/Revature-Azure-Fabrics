```python
import pandas as pd

medical_data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Medicine_Distributed': [1200, 1500, 1800, 1700, 2100, 2500],
    'Hospital_Orders': [30, 45, 50, 48, 60, 75],
    'Delivery_Time': [5, 4, 6, 5, 3, 2],
    'Revenue': [50000, 65000, 72000, 70000, 85000, 95000]
}

df = pd.DataFrame(medical_data)

print(df)
```

1. medicinne distribution over months


```python
import matplotlib.pyplot as plt

plt.plot(df['Month'], df['Medicine_Distributed'],
         marker='o')

plt.title("Medicine Distribution Over Months")
plt.xlabel("Month")
plt.ylabel("Medicine Distributed")
plt.grid(True)

plt.savefig("medicine_distribution_line.png")
plt.show()
```

2.monthly revenue growth

```python
import matplotlib.pyplot as plt

plt.plot(df['Month'], df['Revenue'],
         marker='o')

plt.title("Monthly Revenue Growth")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)

plt.savefig("revenue_growth_line.png")
plt.show()
```

3. multiple line chart

```python
import matplotlib.pyplot as plt

plt.plot(df['Month'],
         df['Medicine_Distributed'],
         marker='o',
         label='Medicine Distributed')

plt.plot(df['Month'],
         df['Hospital_Orders'],
         marker='s',
         label='Hospital Orders')

plt.title("Medicine Distribution vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.grid(True)

plt.savefig("multiple_line_chart.png")
plt.show()
```

5.vertical bar charts for hospital orders

```python
import matplotlib.pyplot as plt

plt.bar(df['Month'],
        df['Hospital_Orders'])

plt.title("Hospital Orders By Month")
plt.xlabel("Month")
plt.ylabel("Orders")

plt.savefig("hospital_orders_bar.png")
plt.show()
```

6.horizantal bar chart for revenue

```python
import matplotlib.pyplot as plt

plt.barh(df['Month'],
         df['Revenue'])

plt.title("Revenue By Month")
plt.xlabel("Revenue")
plt.ylabel("Month")

plt.savefig("revenue_horizontal_bar.png")
plt.show()
```

7. grouped bar chart

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(len(df['Month']))
width = 0.4

plt.bar(x - width/2,
        df['Medicine_Distributed'],
        width,
        label='Medicine Distributed')

plt.bar(x + width/2,
        df['Hospital_Orders'],
        width,
        label='Hospital Orders')

plt.xticks(x, df['Month'])

plt.title("Medicine Distribution vs Hospital Orders")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()

plt.savefig("grouped_bar_chart.png")
plt.show()
```

13. medicine distribution precentage by month

```python
import matplotlib.pyplot as plt

plt.pie(df['Medicine_Distributed'],
        labels=df['Month'],
        autopct='%1.1f%%')

plt.title("Medicine Distribution Percentage")

plt.savefig("medicine_distribution_pie.png")
plt.show()
```

14. revenue contribution by month

```python
import matplotlib.pyplot as plt

plt.pie(df['Revenue'],
        labels=df['Month'],
        autopct='%1.1f%%')

plt.title("Revenue Contribution By Month")

plt.savefig("revenue_contribution_pie.png")
plt.show()
```

15. higlight highest revenue month using explode

```python
import matplotlib.pyplot as plt

highest_index = df['Revenue'].idxmax()

explode = [0, 0, 0, 0, 0, 0]
explode[highest_index] = 0.2

plt.pie(df['Revenue'],
        labels=df['Month'],
        autopct='%1.1f%%',
        explode=explode)

plt.title("Revenue Contribution (Highest Revenue Highlighted)")

plt.savefig("highest_revenue_pie.png")
plt.show()
```
