1. write products into csv file

```python
import csv

products = [
    [101, "Laptop", "Electronics", 75000, 10],
    [102, "Mouse", "Accessories", 800, 50],
    [103, "Keyboard", "Accessories", 1500, 30]
]

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["ID", "Name", "Category", "Price", "Quantity"])

    writer.writerows(products)

print("CSV File Created Successfully")
```

2.read data from csv file

```python
import csv

with open("products.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

3.delete product from csv file

```python
import csv

updated_products = []

with open("products.csv", "r") as file:
    reader = csv.reader(file)

    for row in reader:

        if row[0] == "ID":
            updated_products.append(row)

        elif row[0] != "103":
            updated_products.append(row)

with open("products.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerows(updated_products)

print("Product Deleted Successfully")
```

4.display products with price greater than 5000


```python
import csv

with open("products.csv", "r") as file:
    reader = csv.reader(file)

    next(reader)

    for row in reader:

        if int(row[3]) > 5000:
            print(row)
```

5.write products into json file

```python
import json

products = [
    {"id": 101, "name": "Laptop", "category": "Electronics", "price": 75000, "quantity": 10},
    {"id": 102, "name": "Mouse", "category": "Accessories", "price": 800, "quantity": 50},
    {"id": 103, "name": "Keyboard", "category": "Accessories", "price": 1500, "quantity": 30},
    {"id": 104, "name": "Mobile", "category": "Electronics", "price": 25000, "quantity": 15},
    {"id": 105, "name": "Headphones", "category": "Accessories", "price": 2000, "quantity": 20}
]

with open("products.json", "w") as file:
    json.dump(products, file, indent=4)

print("JSON File Created Successfully")
```

6.read data from json file

```python
import json

with open("products.json", "r") as file:
    data = json.load(file)

    for product in data:
        print(product)
```

7.delete product from json file

```python
import json

with open("products.json", "r") as file:
    data = json.load(file)

updated_data = []

for product in data:

    if product["id"] != 102:
        updated_data.append(product)

with open("products.json", "w") as file:
    json.dump(updated_data, file, indent=4)

print("Product Deleted Successfully")
```

8.update product quality

```python
import json

with open("products.json", "r") as file:
    data = json.load(file)

for product in data:

    if product["name"] == "Laptop":
        product["quantity"] = 25

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)

print("Quantity Updated Successfully")
```

```python
import json

new_product = {
    "id": 106,
    "name": "Monitor",
    "category": "Electronics",
    "price": 12000,
    "quantity": 12
}

with open("products.json", "r") as file:
    data = json.load(file)

data.append(new_product)

with open("products.json", "w") as file:
    json.dump(data, file, indent=4)

print("New Product Added Successfully")
```
