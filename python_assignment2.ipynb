1.array sum


```python
numbers = [45, 12, 78, 34, 78, 90, 12, 56, 23, 45]
def array_sum(arr):
    total = 0

    for i in arr:
        total += i

    return total


print("Array Sum =", array_sum(numbers))
```

2.Maximum element


```python
def find_max(arr):

    largest = arr[0]

    for i in arr:
        if i > largest:
            largest = i

    return largest


print("Maximum Element =", find_max(numbers))
```

3.even numbers counter

```python

def count_even(arr):

    count = 0

    for i in arr:
        if i % 2 == 0:
            count += 1

    return count


print("Even Numbers Count =", count_even(numbers))
```

4.reverse array


```python
def reverse_array(arr):

    return arr[::-1]


print("Reversed Array =", reverse_array(numbers))
```

5.search element and return index


```python
def search_element(arr, key):

    for i in range(len(arr)):

        if arr[i] == key:
            return i

    return -1


print("Index Position =", search_element(numbers, 90))
```

6.palindrome list

```python
def palindrome_list(arr):

    if arr == arr[::-1]:
        return True

    else:
        return False


pal_list = [1, 2, 3, 2, 1]

print("Palindrome =", palindrome_list(pal_list))
```

7.second largest nnumber


```python
def second_largest(arr):

    first = second = -1

    for i in arr:

        if i > first:
            second = first
            first = i

        elif i > second and i != first:
            second = i

    return second


print("Second Largest =", second_largest(numbers))
```

8.duplicate elements


```python
def find_duplicates(arr):

    duplicates = []

    for i in arr:

        if arr.count(i) > 1 and i not in duplicates:
            duplicates.append(i)

    return duplicates


print("Duplicate Elements =", find_duplicates(numbers))
```

9.student marks function

```python
def student_marks(marks):

    total = sum(marks)

    average = total / len(marks)

    highest = marks[0]
    lowest = marks[0]

    for i in marks:

        if i > highest:
            highest = i

        if i < lowest:
            lowest = i

    return average, highest, lowest


marks = [85, 90, 78, 92, 88]

avg, high, low = student_marks(marks)

print("Average =", avg)
print("Highest =", high)
print("Lowest =", low)
```
